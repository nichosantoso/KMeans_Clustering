```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


#dataset with 35 data for each columns (number in range 1-100)
dataset = pd.DataFrame({
    'a':[10,34,62,74,31,68,39,70,12,72,65,72,87,11,21,23,43,52,66,37,34,99,39,80,96,62,47,89,13,27,71,63,83,95,32],
    'b':[67,56,23,1,5,14,97,77,62,83,92,58,32,44,17,37,85,39,12,16,28,36,45,56,94,80,88,32,26,8,61,53,73,65,22]
})

#Using KMeans Library to define cluster and define centroid for each cluster
kmean = KMeans(n_clusters = 5)
kmean.fit(dataset)
label = kmean.predict(dataset)
centroids = kmean.cluster_centers_
```

```python
#plot data before using KMeans Clustering
plot = plt.figure(figsize=(5,5))

colmap = {1 : 'g',2 : 'r',3 : 'y',4 :'b' , 5 : "m"}
plt.scatter(dataset['a'],dataset['b'], color = 'k')

for idx, centroid in enumerate(centroids):
    plt.scatter(*centroid, color = colmap[idx+1])

plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```

```python
#plot data after using KMeans Clustering
plot = plt.figure(figsize=(5,5))

colmap = {1 : 'g',2 : 'r',3 : 'y',4 :'b' , 5 : "m"}
colors = map(lambda x: colmap[x+1],label)
colors1 = list(colors)
plt.scatter(dataset['a'],dataset['b'], color = colors1, alpha = 0.5, edgecolor='k')

for idx, centroid in enumerate(centroids):
    plt.scatter(*centroid, color = colmap[idx+1])

plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```
